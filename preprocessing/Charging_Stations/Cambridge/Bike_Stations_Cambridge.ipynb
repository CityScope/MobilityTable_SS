```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
from pathlib import Path
import geopandas as gpd
from scipy import spatial
```

```python
# Import BlueBikes Stations
filepath_bluebikes_stations = ".././02 Data/BlueBikes/"
bluebikes_stations_df = pd.read_csv(filepath_bluebikes_stations + 'current_bluebikes_stations.csv', header=1)
bluebikes_stations_df.head()
```

```python
bluebikes_stations_cambridge_df=bluebikes_stations_df[bluebikes_stations_df['District']=='Cambridge'].reset_index().drop('index',axis=1)
bluebikes_stations_cambridge_df
```

```python
bluebikes_stations_cambridge_df.Latitude = bluebikes_stations_cambridge_df.Latitude.round(6)
bluebikes_stations_cambridge_df
```

```python
# Save the results of bluebikes stations in Cambridge
bluebikes_path=Path('./Input Database/bluebikes_stations_cambridge.csv')
bluebikes_stations_cambridge_df.to_csv(bluebikes_path,index=False)
```
